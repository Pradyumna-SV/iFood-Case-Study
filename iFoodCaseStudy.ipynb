```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('ml_project1_data.csv')

# Display basic info
print("Basic Info:")
df.info()

# 1. Demographics Analysis
# Convert Year_Birth to Age
df['Age'] = 2024 - df['Year_Birth']  # Assuming current year is 2024

# Distribution of Age
plt.figure(figsize=(8,6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.show()

# Education level distribution (non-encoded)
plt.figure(figsize=(8,6))
sns.countplot(x='Education', data=df)
plt.title('Education Level Distribution')
plt.show()

# Marital Status distribution (non-encoded)
plt.figure(figsize=(8,6))
sns.countplot(x='Marital_Status', data=df)
plt.title('Marital Status Distribution')
plt.show()

# 2. Income Analysis
# Handling missing income data
df['Income'] = df['Income'].fillna(df['Income'].median())  # Impute with median value

# Income distribution
plt.figure(figsize=(8,6))
sns.histplot(df['Income'], kde=True, bins=30)
plt.title('Income Distribution')
plt.show()

from sklearn.cluster import KMeans

# Reshape income column for KMeans (it expects 2D array)
income_data = df[['Income']].dropna()

# Set the number of clusters (bins), for example, 3
kmeans = KMeans(n_clusters=3, random_state=0)
df['Income_Segment_KMeans'] = kmeans.fit_predict(income_data)

# Map the cluster labels to something more meaningful (Low, Medium, High)
df['Income_Segment_KMeans'] = df['Income_Segment_KMeans'].map({0: 'Low', 1: 'Medium', 2: 'High'})

# Distribution of income segments
plt.figure(figsize=(8, 6))
sns.countplot(x='Income_Segment_KMeans', data=df)
plt.title('Income Segmentation Using KMeans Clustering')
plt.show()


# 3. Customer Segmentation
# Kidhome vs Income
plt.figure(figsize=(8,6))
sns.boxplot(x='Kidhome', y='Income', data=df)
plt.title('Income by Number of Kids at Home')
plt.show()

# Teenhome vs Income
plt.figure(figsize=(8,6))
sns.boxplot(x='Teenhome', y='Income', data=df)
plt.title('Income by Number of Teens at Home')
plt.show()

# 4. Purchasing Behavior Analysis
# Correlation of spending on different products
spending_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df_spending = df[spending_columns]
spending_corr = df_spending.corr()

plt.figure(figsize=(8,6))
sns.heatmap(spending_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Spending Correlation Heatmap')
plt.show()

# Purchases by channel
purchase_columns = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df_purchases = df[purchase_columns]
purchase_corr = df_purchases.corr()

plt.figure(figsize=(8,6))
sns.heatmap(purchase_corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Purchasing Channel Correlation Heatmap')
plt.show()

# 5. Campaign Response Analysis
# Analyzing campaign acceptance
campaign_columns = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
df_campaigns = df[campaign_columns]

# Count of accepted campaigns per customer
df['AcceptedCampaigns'] = df_campaigns.sum(axis=1)

# Distribution of accepted campaigns
plt.figure(figsize=(8,6))
sns.countplot(x='AcceptedCampaigns', data=df)
plt.title('Number of Accepted Campaigns per Customer')
plt.show()

# Response column vs accepted campaigns
plt.figure(figsize=(8,6))
sns.countplot(x='Response', data=df)
plt.title('Campaign Response Distribution')
plt.show()

# 6. Recency and Complaint Behavior
# Recency vs Income
plt.figure(figsize=(8,6))
sns.boxplot(x='Recency', y='Income', data=df)
plt.title('Income vs Recency')
plt.show()

# Recency vs Campaign Response
plt.figure(figsize=(8,6))
sns.boxplot(x='Response', y='Recency', data=df)
plt.title('Recency vs Campaign Response')
plt.show()

# Complaints distribution
plt.figure(figsize=(8,6))
sns.countplot(x='Complain', data=df)
plt.title('Complaint Distribution')
plt.show()

# 7. Revenue and Contact Costs
# Z_CostContact vs Income
plt.figure(figsize=(8,6))
sns.boxplot(x='Z_CostContact', y='Income', data=df)
plt.title('Income vs Cost Contact')
plt.show()

# Z_Revenue vs Income
plt.figure(figsize=(8,6))
sns.boxplot(x='Z_Revenue', y='Income', data=df)
plt.title('Income vs Z Revenue')
plt.show()

# 8. Exploratory Data Analysis (EDA)
# Descriptive Statistics for Numerical Columns
print("Descriptive Statistics for Numerical Features:")
print(df.describe())

# Correlation matrix for numerical features
numerical_df = df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# 9. Outlier Detection using Boxplots
# Detecting outliers in spending columns
plt.figure(figsize=(8,6))
sns.boxplot(data=df_spending)
plt.title('Outliers in Spending Columns')
plt.show()

# Detecting outliers in Income column
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Income'])
plt.title('Outliers in Income')
plt.show()

# 10. Behavioral Segmentation
# Segmenting customers based on campaigns, income, and response
df_segmented = df.groupby(['Income_Segment_KMeans', 'Response']).agg({
    'AcceptedCmp1': 'sum',
    'AcceptedCmp2': 'sum',
    'AcceptedCmp3': 'sum',
    'AcceptedCmp4': 'sum',
    'AcceptedCmp5': 'sum',
    'Income': 'mean',
}).reset_index()

# Visualize Segmentation
plt.figure(figsize=(10,8))
sns.barplot(x='Income_Segment_KMeans', y='Income', hue='Response', data=df_segmented)
plt.title('Income by Segment and Campaign Response')
plt.show()

# 11. Encoding for Correlation Matrix
# Encoding categorical columns
le = LabelEncoder()
encoded_df = df.copy()

# Encoding categorical columns (Education, Marital_Status)
encoded_df['Education'] = le.fit_transform(df['Education'])
encoded_df['Marital_Status'] = le.fit_transform(df['Marital_Status'])

# Correlation matrix for encoded data
encoded_numerical_df = encoded_df.select_dtypes(include=[np.number])
plt.figure(figsize=(12, 8))
sns.heatmap(encoded_numerical_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Encoded Features')
plt.show()

```

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('ml_project1_data.csv')

# Convert 'Dt_Customer' to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# Assuming the 'Recency' column is in days, let's calculate the current date
current_date = df['Dt_Customer'].max() + pd.DateOffset(days=1)  # Assuming the last date is the most recent

# 1. Recency: Days since last purchase
df['Recency'] = (current_date - df['Dt_Customer']).dt.days

# 2. Frequency: Total purchases made (sum of NumDealsPurchases, NumWebPurchases, NumCatalogPurchases, NumStorePurchases)
df['Frequency'] = df[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']].sum(axis=1)

# 3. Monetary: Total spent (sum of MntWines, MntFruits, MntMeatProducts, MntFishProducts, MntSweetProducts, MntGoldProds)
df['Monetary'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum(axis=1)

# Create a new DataFrame to avoid modifying the original one directly
rfm = df[['ID', 'Recency', 'Frequency', 'Monetary']].copy()

# 4. Perform RFM Segmentation (using quartiles to define 'High', 'Medium', 'Low' categories)
rfm['Recency_Segment'] = pd.qcut(rfm['Recency'], 3, labels=['High', 'Medium', 'Low'])
rfm['Frequency_Segment'] = pd.qcut(rfm['Frequency'], 3, labels=['Low', 'Medium', 'High'])
rfm['Monetary_Segment'] = pd.qcut(rfm['Monetary'], 3, labels=['Low', 'Medium', 'High'])

# 5. Calculate RFM Score: Combine the segments into a single RFM score
rfm['RFM_Score'] = rfm['Recency_Segment'].astype(str) + rfm['Frequency_Segment'].astype(str) + rfm['Monetary_Segment'].astype(str)

# Display the RFM dataframe
print(rfm.head())

# Insights Generation:

# Insight 1: Distribution of RFM Segments
rfm_segments = rfm['RFM_Score'].value_counts()
print("\nRFM Segment Distribution:")
print(rfm_segments)

# Insight 2: Top 10 Customers Based on Monetary Value
top_10_customers = rfm.sort_values(by='Monetary', ascending=False).head(10)
print("\nTop 10 Customers based on Monetary:")
print(top_10_customers[['ID', 'Monetary']])

# Insight 3: Recency vs Frequency
recency_vs_frequency = rfm.groupby(['Recency_Segment', 'Frequency_Segment']).size().unstack(fill_value=0)
print("\nRecency vs Frequency Distribution:")
print(recency_vs_frequency)

# Insight 4: High-Value Customers (High Frequency & High Monetary)
high_value_customers = rfm.loc[(rfm['Frequency_Segment'] == 'High') & (rfm['Monetary_Segment'] == 'High')]
print("\nHigh-Value Customers (High Frequency & High Monetary):")
print(high_value_customers[['ID', 'Frequency', 'Monetary']])

# Insight 5: Customers with Low Recency but High Frequency (Potential Re-engagement Targets)
low_recency_high_frequency = rfm.loc[(rfm['Recency_Segment'] == 'Low') & (rfm['Frequency_Segment'] == 'High')]
print("\nCustomers with Low Recency but High Frequency (Re-engagement Targets):")
print(low_recency_high_frequency[['ID', 'Recency', 'Frequency']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('ml_project1_data.csv')

# Function to create Pareto chart and generate insights for a column
def create_pareto_chart_with_insights(data, column_name):
    # Calculate frequency of the column values
    value_counts = data[column_name].value_counts()

    # Calculate the cumulative percentage
    cumsum = value_counts.cumsum() / value_counts.sum() * 100

    # Plotting the Pareto chart
    fig, ax = plt.subplots(figsize=(10, 6))

    # Bar chart for frequencies
    ax.bar(value_counts.index, value_counts.values, color='skyblue')

    # Line chart for cumulative percentage
    ax2 = ax.twinx()
    ax2.plot(value_counts.index, cumsum, color='red', marker='D', linestyle='-', linewidth=2)

    # Titles and labels
    ax.set_title(f'Pareto Chart for {column_name}', fontsize=14)
    ax.set_xlabel(column_name, fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)
    ax2.set_ylabel('Cumulative Percentage', fontsize=12)

    # Display the plot
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    # Insights
    print(f"\nInsights for {column_name}:")
    print(f"Total categories: {len(value_counts)}")
    print(f"Most frequent category: {value_counts.idxmax()} with {value_counts.max()} occurrences")
    print(f"Categories contributing to 80% of the total:")
    
    # Identify categories that contribute to 80%
    cumulative_threshold = 80
    significant_categories = value_counts[cumsum <= cumulative_threshold]
    print(f"  - {significant_categories.index.tolist()}")

    # Total count of top categories contributing to 80%
    total_significant = significant_categories.sum()
    print(f"Total count of significant categories: {total_significant}")
    print(f"Cumulative percentage at 80% threshold: {cumsum[significant_categories.index[-1]]:.2f}%\n")

# Iterate over all columns and create Pareto chart with insights for categorical columns
for column in df.columns:
    if df[column].dtype == 'object':  # Check if the column is categorical (object type)
        create_pareto_chart_with_insights(df, column)
```
